```python
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
plt.plot(x,y,color='black',marker='+')
plt.show()
```

```python
import numpy as np
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color='black',linewidth=5,label="x^3")
plt.plot(x,y2,color='red',linewidth=5,label="x^2")
plt.axis([0,10,0,200])
plt.title("My-demo")
plt.grid()
plt.text(2,80,'Text',fontsize=12)
plt.ylabel("x^3")
plt.xlabel("x")
plt.legend()
plt.show()
```

```python
#pie graphs
labels=["A","B","C","D"]
sizes=[3,4,6,2]
colors=["blue","purple","red","pink"]
explode=[0.1,0.2,0,0]
plt.title("split among classes")
plt.pie(sizes,colors=colors,labels=labels,explode=explode,autopct="%.2f%%",startangle=100)
plt.axis("equal")
plt.show()
```
