```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
# Example dataset
tips = sns.load_dataset("tips")

# Visualizing the data using relplot
sns.relplot(data=tips, x='total_bill', y='tip', hue='smoker',style='smoker',size='size')

# Show the plot
plt.show()
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
california_housing = fetch_california_housing()
# Create a DataFrame from the dataset
house = pd.DataFrame(california_housing.data, columns=california_housing.feature_names)
# Add the target 'PRICE' column to the DataFrame
house['PRICE'] = california_housing.target
# Display the first few rows of the DataFrame
print(house.head())
```

```python
#DISTRIBUTION PLOT 
sns.distplot(house['PRICE'])
```

```python
#correlation as -ve and +ve correlation
#how to use heat map
correlation=house.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, 
            cbar=True, 
            square=True, 
            fmt='.1f', 
            annot=True, 
            annot_kws={'size': 8}, 
            cmap='Blues')
plt.show()
```
