```python
import numpy as np

vector_1 = np.asarray([2, 3])
vector_2 = np.asarray([3, -2])

dot = np.dot(vector_1, vector_2)  # Dot product is valid for any same-dimensional vectors

# Cross product is only for 3D vectors, so let's convert them to 3D (by adding a zero z-component)
vector_1_3d = np.array([2, 3, 0])
vector_2_3d = np.array([3, -2, 0])
cross = np.cross(vector_1_3d, vector_2_3d)  # Compute cross product in 3D

print("Dot Product:", dot)
print("Cross Product:", cross)
```

```python
import numpy as np

# Define vectors A and B
A = np.array([3, 4, 5])
B = np.array([1, 2, 3])

# Compute the dot product A . B
dot_product = np.dot(A, B)

# Compute the magnitude squared of B
B_magnitude_squared = np.dot(B, B)

# Compute the projection
projection = (dot_product / B_magnitude_squared) * B

print("Projection of A onto B:", projection)
```
